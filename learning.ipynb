```python
x = torch.arange(1., 10.)
x, x.shape
```

```python
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(9, 1)
x_reshaped, x_reshaped.shape
```

```python
z = x.view(1, 9)
# View is similar to reshape, but it can handle non-contiguous tensors
# View is faster than reshape
# View is more memory efficient than reshape
# View is more flexible than reshape
# View is more intuitive than reshape
# View is more readable than reshape
# View is more pythonic than reshape
z[:,0]=33
z, x
```

```python
# stack tensors on top of each other
x_stacked = torch.stack([x,x,x,x, x ], dim=1)
x_stacked, x_stacked.shape
```
