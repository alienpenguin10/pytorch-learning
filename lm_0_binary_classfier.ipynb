```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
inputs = torch.randint(low=0, high=100, size=(10,2), dtype=torch.float32)
print(inputs)
```

```python
labels = torch.randint(low=0, high=2, size=(10,)) #size=(10,) is not same as size=(10,1)
print(labels)
```

```python
model = nn.Sequential(
    nn.Linear(inputs.shape[1], 100),
    nn.Sigmoid(),
    nn.Linear(100, labels.shape[1]),
    nn.Sigmoid()
)
optimizer = optim.SGD(model.parameters(), lr=0.01)
criterion = nn.BCELoss()
```

```python
for step in range(1000):
    optimizer.zero_grad() #multiple grad computation before update; no needed;
    loss = criterion(model(inputs), labels)
    loss.backward()
    optimizer.step()
```
